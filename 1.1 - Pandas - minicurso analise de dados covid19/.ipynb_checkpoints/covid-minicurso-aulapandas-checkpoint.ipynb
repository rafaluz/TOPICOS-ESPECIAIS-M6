```python
# Carregando os dados
dados = pd.read_csv('HIST_PAINEL_COVIDBR_04nov2020.csv', delimiter=';')
```

```python
# Filtrando estado do PI
PIFull = dados[dados.estado == 'PI']
```

```python
PIFull = PIFull[PIFull.municipio.isna() & PIFull.populacaoTCU2019.notna()]
```

```python
PI = PIFull[['data','casosAcumulado','obitosAcumulado','casosNovos','obitosNovos']].groupby(['data']).sum()
```

```python
PI.index = range(len(PI))
```

```python
def prediction(X_train, X_test, y_train,degree):

    # Criando o modelo
    modelo_polinomial = make_pipeline(PolynomialFeatures(degree), Ridge())
    # Realizando o treinamento
    modelo_polinomial.fit(X_train, y_train)


    # Realizando a predio
    pred_values = modelo_polinomial.predict(X_test)
    return pred_values
```

```python
def pred(df,column):
    
    # Definindo X e Y
    X = df.index.values
    y = df[column].values
    
    # Dividindo o DataFrame em janelas deslizantes para treino e teste
    # X_train = X[0:200].reshape(-1,1)
    # X_test = X[200:].reshape(-1,1)
    
    # y_train = y[0:200].reshape(-1,1)
    # y_test = y[200:].reshape(-1,1)


    ### Descomentar para usar o tran_test_split (comentar as janelas deslizantes)
    X = X.reshape(-1,1)
    y = y.reshape(-1,1)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=False, shuffle=False )

    # Realizando a predio
    pred_values = prediction(X_train,X_test,y_train,1)
    
    # Imprimindo o resultado
    print(explained_variance_score(y_test,pred_values))
    
    # Plotando o grfico
    plt.plot(X_train,y_train,color='red')
    plt.plot(X_test,y_test,color='pink')
    plt.plot(X_test,pred_values,color='black')
    plt.show()
```

```python
pred(PI,'obitosAcumulado')
```
