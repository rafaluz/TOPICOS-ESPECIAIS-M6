## Importação de base de dados com mais ações

Agora que iremos utilizar várias ações, precisamos de uma referência de desempenho, por isso iremos comparar as ações com a BOVA11.
- BOVA11: https://www.sunoresearch.com.br/artigos/bova11/

```python
#definindo as ações que serão utilizadas
acoes = ['GOLL4.SA', 'CVCB3.SA', 'WEGE3.SA', 'MGLU3.SA', 'TOTS3.SA', 'BOVA11.SA']
```

```python
# criando o dataframe com as ações
acoes_df = pd.DataFrame()
for acao in acoes:
  acoes_df[acao] = yf.download(acao, start='2015-01-01')['Close']
```

```python
# renomentado as colunas para facilitar as chamadas
acoes_df = acoes_df.rename(columns={'GOLL4.SA': 'GOL', 'CVCB3.SA': 'CVC', 'WEGE3.SA': 'WEGE',
                                    'MGLU3.SA': 'MGLU', 'TOTS3.SA': 'TOTS', 'BOVA11.SA': 'BOVA'})
```

```python
#exibindo as colunas
acoes_df.columns[0:]
```

```python
acoes_df.head()
```

```python
# verificando a presença de registros nulos
acoes_df.isnull().sum()
```

```python
acoes_df.shape
```

```python
# removendo os registros nulos
acoes_df.dropna(inplace=True)
```

```python
acoes_df.shape
```

```python
acoes_df.isnull().sum()
```

```python
# salvando nossos dados
acoes_df.to_csv('acoes.csv')
```

## Visualização dos dados

### Histograma

Um histograma é uma espécie de gráfico de barras que demonstra uma distribuição de frequências. No histograma, a base de cada uma das barras representa uma classe e a altura representa a quantidade ou frequência absoluta com que o valor de cada classe ocorre.

```python
sns.histplot(acoes_df['GOL']);
```

```python
len(acoes_df.columns)
```

```python
acoes_df.columns
```

```python
np.arange(1, len(acoes_df.columns))
```

```python
# exibindo plotagens dos histogramas de cada ações
plt.figure(figsize=(10,50))
for i in np.arange(1, len(acoes_df.columns)):
  plt.subplot(7, 1, i)
  sns.histplot(acoes_df[acoes_df.columns[i]], kde = True)
  plt.title(acoes_df.columns[i])
```

### Diagrama de Caixa 



Em estatística descritiva, diagrama de caixa, diagrama de extremos e quartis, boxplot ou box plot é uma ferramenta gráfica para representar a variação de dados observados de uma variável numérica por meio de quartis. 

É um método alternativo ao histograma e ao ramo-e-folha para representar os dados, fornecendo informação sobre as seguintes características do conjunto de dados: localização, dispersão, assimetria, comprimento da cauda e outliers (medidas discrepantes).

```python
acoes_df['GOL'].describe()
```

```python
#exibindo o boxplot
sns.boxplot(x = acoes_df['GOL']);
```

```python
# exibindo o boxplot para todas as ações
plt.figure(figsize=(10,50))
for i in np.arange(1, len(acoes_df.columns)):
  plt.subplot(7, 1, i)
  sns.boxplot(x = acoes_df[acoes_df.columns[i]])
  plt.title(acoes_df.columns[i])
```

### Gráfico de linha

O gráfico de linha é um tipo de gráfico que exibe informações com uma série de pontos de dados chamados de marcadores ligados por segmentos de linha reta. É um tipo básico de gráfico comum em muitos campos. 

```python
acoes_df.plot(x = 'Date', figsize = (15,7), title = 'Histórico do preço das ações');
```

```python
acoes_df
```

### Normalizando os dados

```python
acoes_df_normalizado = acoes_df.copy()
for i in acoes_df_normalizado.columns[1:]:
  acoes_df_normalizado[i] = acoes_df_normalizado[i] / acoes_df_normalizado[i][0]
```

```python
acoes_df_normalizado
```

```python
acoes_df_normalizado.iloc[:,:5]
```

```python
# exibindo o gráfico de linhas com os dados normalizados
acoes_df_normalizado.plot(x = 'Date', figsize = (15,7), title = 'Histórico do preço das ações - normalizado');
```

```python
# exibindo sem a MGLU para visualizar melhor as outras ações
# acoes_df_normalizado.drop('MGLU', axis = 1 ).plot(x = 'Date', figsize = (15,7), title = 'Histórico do preço das ações - normalizado');
```

## Gráficos Dinâmicos com plotly.express 

```python
# gráfico das ações
figura = px.line(title = 'Histórico do preço das ações', height=500, width=1000)
for i in acoes_df.columns[1:]:
  figura.add_scatter(x = acoes_df['Date'], y = acoes_df[i], name = i)
figura.show()
```

```python
# gráfico das ações normalizado
figura = px.line(title = 'Histórico do preço das ações - normalizado', height=500, width=1000)
for i in acoes_df_normalizado.columns[1:]:
  figura.add_scatter(x = acoes_df_normalizado['Date'], y = acoes_df_normalizado[i], name = i)
figura.show()
```
