### Pegando e Plotando as imagens originais

```python
#Pegar as imagens da pasta
files = glob.glob('bg1-bg10/*.tif')
imagens = imread_collection(files)
```

```python
#Plotando imagens originais
fig, ax = plt.subplots(1,len(imagens),figsize=(15, 15))
for i,imagem in enumerate(imagens):
    ax[i].imshow(imagem, cmap='gray')
```
