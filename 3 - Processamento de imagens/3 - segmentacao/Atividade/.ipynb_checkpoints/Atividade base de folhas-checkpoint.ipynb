```python
#Importações
from skimage.io import imread_collection
import glob
import numpy as np
from skimage.io import imread,imsave
from skimage.color import rgb2gray
from skimage.exposure import histogram
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from skimage.filters import sobel,roberts,prewitt,scharr
from skimage.filters import threshold_otsu
%matplotlib inline
```

### Pegando e Plotando as imagens originais

```python
#Pegar as imagens da pasta
files = glob.glob('bg1-bg10/*.tif')
imagens = imread_collection(files)
```

```python
#Plotando imagens originais
fig, ax = plt.subplots(1,10,figsize=(15, 15))
for i,imagem in enumerate(imagens):
    ax[i].imshow(imagem, cmap='gray')
```

###  1. Aplique o algoritmo K-means utilizando os pixels como pesos, após isso utilize algum detector de borda;

```python
# Aplicando Kmeans
kimagens = []
for imagem in imagens:
    model = KMeans(n_clusters=2, random_state=0).fit(np.reshape(imagem,(-1, 1)))
    segmented = model.predict(np.reshape(imagem,(-1, 1)))
    segmented = np.reshape(segmented,(imagem.shape[0],imagem.shape[1]))
    kimagens.append(segmented)

fig, ax = plt.subplots(1,10,figsize=(15, 15))
for i,imagem in enumerate(kimagens):
    ax[i].imshow(imagem, cmap='gray')
```

```python
#Função de padronização que identifica se tem mais pixels brancos ou pretos na imagem e inverte nas folhas que ficaram pretas
def altera_black_and_white(imagens):
    imagens_black_and_white = []
    for i, imagem in enumerate(imagens):    
        pixel_preto = 0
        pixel_branco = 0
        for linha in range(imagem.shape[0]):
            for coluna in range(imagem.shape[1]):
                if imagem[linha][coluna] == 0:
                    pixel_preto += 1
                else:
                    pixel_branco += 1   
        if pixel_preto > pixel_branco: 
            np.place(imagem, imagem == 0, [255])
            imagens_black_and_white.append(imagem)
        else:
            imagens_black_and_white.append(imagem)
    return imagens_black_and_white
```

```python
#Chamada da função e exibição das novas imagens com cores invertidas
kimagens_novas = altera_black_and_white(kimagens)
fig, ax = plt.subplots(1,10,figsize=(15, 15))
for i,imagem in enumerate(kimagens_novas):
    ax[i].imshow(imagem, cmap='gray')
```

```python
#Aplicando borda sobel
fig, ax = plt.subplots(1,10,figsize=(15, 15))
for i,imagem in enumerate(kimagens_novas):
    ax[i].imshow(sobel(imagem), cmap='gray')
```

### 2. Aplique o limiar de Otsu, após isso utilize o mesmo detector de borda do item anterior;

```python
#Aplicando otsu
otzuimagens = []
for imagem in imagens:
    cinza = rgb2gray(imagem)
    #threshold_otsu
    otsu = threshold_otsu(cinza)
    binary_otsu = cinza < otsu 
    otzuimagens.append(binary_otsu)

fig, ax = plt.subplots(1,10,figsize=(15, 15))
for i,imagem in enumerate(otzuimagens):
    ax[i].imshow(imagem, cmap='gray')
```

```python
#Aplicando borda sobel
fig, ax = plt.subplots(1,10,figsize=(15, 15))
for i,imagem in enumerate(otzuimagens):
    ax[i].imshow(sobel(imagem), cmap='gray')
```
