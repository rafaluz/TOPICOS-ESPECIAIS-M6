# Descritores

Em processamento de imagem e visão computacional, o termo "descritor" refere-se a um conjunto de características ou informações que são extraídas de uma imagem para descrever suas propriedades específicas. Esses descritores são utilizados para representar de forma mais compacta e significativa o conteúdo da imagem, facilitando a análise, comparação e interpretação por algoritmos de computador.

Existem diferentes tipos de descritores, cada um projetado para capturar aspectos específicos da informação visual em uma imagem. Alguns exemplos comuns de descritores incluem:

## Descritores de Cor:
Capturam informações sobre as cores presentes na imagem. Exemplos incluem histogramas de cor, médias de cor e momentos de cor.

## Descritores de Textura:
Representam as características texturais da imagem, como rugosidade, suavidade, etc. Exemplos incluem matriz de coocorrência, descritores de transformada de Fourier e Histograma de Gradientes Orientados (HOG).

## Descritores de Forma:
Descrevem as características de forma dos objetos na imagem. Exemplos incluem momentos de Hu, descritores elípticos e descritores baseados em contornos.

## Descritores Locais:
Concentram-se em regiões específicas da imagem. Exemplos incluem SIFT (Scale-Invariant Feature Transform), SURF (Speeded-Up Robust Features) e ORB (Oriented FAST and Rotated BRIEF).

## Descritores de Estrutura:
Descrevem padrões espaciais e relacionamentos entre diferentes partes da imagem. Exemplos incluem Matriz de Coocorrência de Níveis de Cinza (GLCM) e Matriz de Comprimentos de Execução de Níveis de Cinza (GLRLM).

Esses descritores são fundamentais em várias aplicações, como reconhecimento de objetos, classificação de imagens, detecção de padrões, indexação de imagens, entre outras. A escolha do descritor depende da tarefa específica que está sendo realizada e das características que são mais relevantes para a análise desejada.

Nesta aula iremos utilizar o descritor HOG como demonstração.

### HOG
* O algoritmo Histograma de Gradientes Orientados (Histograms of
Oriented Gradients - HOG) é um descritor que calcula o
histograma da orientação dos gradientes na imagem;
* O descritor final é um vetor de histogramas extraídos da imagem;
* O algoritmo se baseia na ideia que a forma e a aparência de um
objeto podem ser descritas muitas vezes pela intensidade dos
gradientes ou a direção das bordas.

```python
import matplotlib.pyplot as plt

from skimage.feature import hog
from skimage import data, exposure

from skimage.io import imread, imread_collection
import glob
import time
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import cohen_kappa_score
%matplotlib inline
```

```python
fd, hog_image = hog(image, orientations=8, pixels_per_cell=(3, 3),
                        cells_per_block=(1,1), visualize=True, channel_axis=2) #, multichannel=True
```

## Aplicando o descritor HOG em toda a base de imagens

```python
# ------------------- GLAUCOMA - RIMONE-DB-R2
#Pasta raiz
path_database = 'glaucoma/'
#Pastas das classes (maligno e benigno) / NÃO SE ESQUEÇA DE ADEQUAR A EXTENSÃO DO ARQUIVO
imagens_maligno = imread_collection(path_database+'maligno2/*.jpg')
imagens_benigno = imread_collection(path_database+'benigno2/*.jpg')
# ------------------- MELANOMA - MEDNODE
# #Pasta raiz
# path_database = 'mednode/'
# #Pastas das classes (maligno e benigno) / NÃO SE ESQUEÇA DE ADEQUAR A EXTENSÃO DO ARQUIVO
# imagens_maligno = imread_collection(path_database+'melanoma/*.jpeg')
# imagens_benigno = imread_collection(path_database+'notmelanoma/*.jpeg')


# -------------------
#Dois principais parametros do HOG
#Numero de celulas (obs.: enquanto menor, maior sera seu vetor de caracteristicas)
cell_size = (16, 16)
#Numero de celulas do bloco
block_size = (1, 1)

#Nome do arquivo csv que será salvo para a combinação de parametros
nome_do_csv = 'hog_celula'+str(cell_size)+'_bloco'+str(block_size)+'.csv'


#------------------------------------
inicio = time.time()

tamanho_maligno = len(imagens_maligno)
tamanho_benigno = len(imagens_benigno)
tam = tamanho_maligno+tamanho_benigno

matriz = []
target = []

#Aplicacao na base maligna
for i,image in enumerate(imagens_maligno):
    fd, hog_image = hog(image, orientations=8, pixels_per_cell=cell_size,
                        cells_per_block=block_size, visualize=True, channel_axis=2) #, multichannel=True
   
    matriz.append(fd)
    target.append(1)
    
print("terminou maligno")
# Aplicacao na base benigna
for i,image in enumerate(imagens_benigno,start=tamanho_maligno):
    fd, hog_image = hog(image, orientations=8, pixels_per_cell=cell_size,
                        cells_per_block=block_size, visualize=True, channel_axis=2) #, multichannel=True

    matriz.append(fd)
    target.append(0)
  
print("terminou benigno")

# converter lista para np.array
x = np.array(matriz)
target = np.array(target)

#Juntar o X com o Y
xy = np.zeros((x.shape[0],x.shape[1]+1))
xy[:,:-1] = matriz
xy[:,-1] = target.reshape(-1)

# salvando em csv
np.savetxt(nome_do_csv,xy,delimiter=",")
print("arquivo csv salvo!")
fim = time.time()
tempo_processamento = fim-inicio
print("Terminado!")
print(tempo_processamento)       
```

## Carregando o CSV

```python
#Favor carregar o csv com o np, pq se usar o pandas ta bugando 
data = np.genfromtxt(nome_do_csv,delimiter=',')
```

```python
# shape
data.shape
```

### Divisão treino/teste

```python
X = data[:,:-1]
Y = y = data[:,-1]
X_treino, X_teste, Y_treino, Y_teste = train_test_split(X, Y, test_size=0.3, shuffle=True, stratify=Y)
```

### Treinando a MLP

```python
MLP = MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=500, random_state=42)

MLP.fit(X_treino,Y_treino)
```

### Fazendo as predições e calculando a taxa de acerto

```python
valores_preditos_teste = MLP.predict(X_teste)
valores_preditos_treinamento = MLP.predict(X_treino)

acuracia_teste = accuracy_score(Y_teste,
                                valores_preditos_teste)

acuracia_treinamento = accuracy_score(Y_treino,
                                      valores_preditos_treinamento)

kappa_teste = cohen_kappa_score(Y_teste,
                                      valores_preditos_teste)

kappa_treinamento = cohen_kappa_score(Y_treino,
                                      valores_preditos_treinamento)

matriz_confusao_teste = confusion_matrix(Y_teste,
                                         valores_preditos_teste)

print('Acuracia treino = ', acuracia_treinamento)
print('Acuracia teste = ', acuracia_teste)
print('Kappa treino = ', kappa_treinamento)
print('Kappa teste = ', kappa_teste)
print(matriz_confusao_teste)
```

```python
import seaborn as sns
plt.figure(figsize=(8, 6))
sns.set(font_scale=1.4)  # Ajustar o tamanho da fonte

# Criar o heatmap
sns.heatmap(matriz_confusao_teste, annot=True, fmt=".2f", cmap="Blues",
            xticklabels=["Maligno", "Beningno"],
            yticklabels=["Maligno", "Beningno"])

# Adicionar rótulos e título
plt.title("Matriz de Confusão")
plt.xlabel("Classe Real")
plt.ylabel("Classe Predita")

plt.show()
```

## Utilizando o modelo RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
RFC = RandomForestClassifier(max_depth=None, random_state=50, n_estimators=100)
RFC.fit(X, Y)


valores_preditos_teste_RFC = RFC.predict(X_teste)
valores_preditos_treinamento_RFC = RFC.predict(X_treino)

acuracia_teste_RFC = accuracy_score(Y_teste,
                                valores_preditos_teste_RFC)

acuracia_treinamento_RFC = accuracy_score(Y_treino,
                                      valores_preditos_treinamento_RFC)

kappa_teste_RFC = cohen_kappa_score(Y_teste,
                                      valores_preditos_teste_RFC)

kappa_treinamento_RFC = cohen_kappa_score(Y_treino,
                                      valores_preditos_treinamento_RFC)

matriz_confusao_teste_RFC = confusion_matrix(Y_teste,
                                         valores_preditos_teste_RFC)

print('Acuracia treino = ', acuracia_treinamento_RFC)
print('Acuracia teste = ', acuracia_teste_RFC)
print('Kappa treino = ', kappa_treinamento_RFC)
print('Kappa teste = ', kappa_teste_RFC)
print(matriz_confusao_teste_RFC)
```

```python
import seaborn as sns
plt.figure(figsize=(8, 6))
sns.set(font_scale=1.4)  # Ajustar o tamanho da fonte

# Criar o heatmap
sns.heatmap(matriz_confusao_teste_RFC, annot=True, fmt=".2f", cmap="Blues",
            xticklabels=["Maligno", "Beningno"],
            yticklabels=["Maligno", "Beningno"])

# Adicionar rótulos e título
plt.title("Matriz de Confusão")
plt.xlabel("Classe Real")
plt.ylabel("Classe Predita")

plt.show()
```
